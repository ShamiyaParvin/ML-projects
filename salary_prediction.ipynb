```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/Salary_Data.csv')
df
```

```python
import matplotlib.pyplot as plt
x=df['YearsExperience'].tolist()
y=df['Salary'].tolist()
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary-Year Experience Graph')
plt.scatter(x,y)
```

```python
# seperate x
x=df.iloc[:,:-1]
x
```

```python
# seperate y
y=df.iloc[:,-1]
y
```

```python
# training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
# Here no need to do normalisation becoz we need exact value as the prediction answer
# in regression model no normalization
```

```python
# model creation
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
x_test
```

```python
y_test
```

```python
# To plot the graph
import matplotlib.pyplot as plt
plt.scatter(x_train,y_train)
plt.plot(x_test,y_pred,color='red')
```

```python
print(model.predict([[8.5]]))
```

```python
# Performance measurement
# MAE - Mean Absolute Error
from sklearn.metrics import mean_absolute_error
print("Error is",mean_absolute_error(y_test,y_pred))
```

```python
# MAPE - Mean Absolute Percentage Error
from sklearn.metrics import mean_absolute_percentage_error
print("Percentage Error is",mean_absolute_percentage_error(y_test,y_pred))
```

```python
# MSE - Mean Squared Error
from sklearn.metrics import mean_squared_error
data=mean_squared_error(y_test,y_pred)
print("Squared Error is",data)
```

```python
# There is no function to calculate RMSE(Root Mean Squared Error)
sq=np.sqrt(data)
print("Root Mean Squared Error is",sq)
```

```python
# R2 Score ==> Coefficient of determination
from sklearn.metrics import r2_score
print("R2_Score",r2_score(y_test,y_pred))
```
